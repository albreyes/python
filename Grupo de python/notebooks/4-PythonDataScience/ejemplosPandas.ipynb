##**Dataframes**


Una buena opción para empezar un dataframe es usando un diccionario

```
datos = {
    'manzanas': [3, 2, 0, 1],
    'naranjas': [0, 3, 7, 2]
}
datos
```

Solo pasamos el diccionario como argumento a la función dataFrame de pandas

```
compras = pd.DataFrame(datos)

compras
```

Podemos cambiar los índices automáticos con nombres para hacer más expresivo el DataFrame

```
compras = pd.DataFrame(datos, index=['Arturo', 'Eduardo', 'Saul', 'Eloy'])

compras
```

Si quisieramos localizar las compras de Eduardo usamos el atributo loc

```
compras.loc['Eduardo']
```

Si quisieramos generar un dataFrame de 2500 filas x 8 columnas con enteros aleatorios entre 0 y 3 podemos usar la función randInt del paquete random de numpy y después pasarlo convertirlo a DataFrame. Ahí mismo podemos etiquetar las columnas usando una lista de caracteres.

```
import numpy as np
arregloAleatorio=np.random.randint(0, 3, size=(2500, 8))
data = pd.DataFrame(arregloAleatorio, columns=list('ABCDEFGH'))
print(data)
```

Como son muchos datos, podemos visualizar los primeros 5 registros de arriba con la función head() del DataFrame

```
data.head()
```

O los últimos con la función tail()

```
data.tail()
```

Igual se pueden hacer operaciones artiméticas con los datos de diferentes columnas (por ejemplo, sumarlos o multiplicarlos) y asignar el resultado a otra columna.

```
data['A'] = data['B'] + data['C']
data['H'] = data['G'] - data['A']
data['E'] = data['F']
data.head()
```

Si se nos olvida la forma del DataFrame (filas,columnas) podemos usar el atributo shape.

```
data.shape
```

La función info() nos da información de la tabla de datos, como la cuenta de nulos y el tipo de dato.

```
data.info()
```

###**Manejo de duplicados**

Para demostrar el manejo de datos duplicados usemos una tabla mas chica y dupliquemos sus filas. Para eso usaremos la función append. Volvamos con la tabla de compras.

```
temporal=compras.append(compras)
temporal
```

Ahora eliminemos los registros duplicados para dejarlo como estaba.

```
temporal = temporal.drop_duplicates()

temporal
```

Otro argumento importante para drop_duplicates() es *keep*, que tiene tres opciones posibles:

1.   *First*: (predeterminado) Elimina duplicados excepto la primera ocurrencia.
2.   *Last*: elimina duplicados excepto la última ocurrencia.
3. *False*: elimine todos los duplicados.

Como no definimos el argumento keep en el ejemplo anterior, el valor predeterminado era *First*. Esto significa que si dos filas son iguales, pandas eliminará la segunda fila y mantendrán la primera. Usar *Last* tiene el efecto contrario: se elimina la primera fila.

*keep*, por otro lado, eliminará todos los duplicados. Si dos filas son iguales, ambas se eliminarán. Mira lo que sucede con temporal:

```
temporal=temporal.append(temporal)
temporal.drop_duplicates(inplace=True, keep=False)

temporal
```

O sea que no mantuvo ninguno de los duplicados! Eliminó todos
 los registros!

Muchas veces los conjuntos de datos tendrán nombres de columnas detallados con símbolos, palabras en mayúsculas y minúsculas, espacios y errores tipográficos. Para facilitar la selección de datos por nombre de columna, podemos limpiar sus nombres.

```
compras.rename(columns={
        'manzanas': 'Manzanas',
        'naranjas': 'Naranjas_dulces'
    }, inplace=True)


compras.columns
```

###**Valores faltantes**

Al explorar datos, lo más probable es que encuentre valores nulos o faltantes, que son esencialmente marcadores de posición para valores inexistentes. Lo más común es que veas *None* de Python o *np.nan* de NumPy, cada uno de los cuales se maneja de manera diferente en algunas situaciones.

Hay dos opciones para tratar con nulos:

1.   Deshacerse de filas o columnas con valores nulos
2.   Reemplazar valores nulos con valores no nulos, una técnica conocida como imputación







```
movies_df = pd.read_csv("https://raw.githubusercontent.com/LearnDataSci/articles/master/Python%20Pandas%20Tutorial%20A%20Complete%20Introduction%20for%20Beginners/IMDB-Movie-Data.csv", index_col="Title")
movies_df
```

Calculemos el número total de nulos en cada columna de nuestro conjunto de datos. El primer paso es comprobar qué celdas de nuestro DataFrame son nulas:

```
movies_df.isnull()
```

Note que isnull() devuelve un DataFrame donde cada celda es Verdadero o Falso dependiendo del estado nulo de esa celda.

Para contar el número de valores nulos en cada columna utilizamos una función agregada para sumar:

```
movies_df.isnull().sum()
```

.isnull() por sí solo no es muy útil y generalmente se usa junto con otros métodos, como sum().

Ahora podemos ver que a nuestros datos les faltan 128 valores para Revenue (Millions) y 64 valores faltantes para Metascore.

###**Remoción de valores nulos**

Los científicos y analistas de datos se enfrentan regularmente al dilema de descartar o imputar valores nulos, y es una decisión que requiere un conocimiento profundo de sus datos y su contexto. En general, solo se sugiere eliminar datos nulos si falta una pequeña cantidad de datos.

Eliminar valores nulos es bastante simple:

```
movies_df.dropna()
```

Esta operación eliminará cualquier fila con al menos un único valor nulo, pero devolverá un nuevo DataFrame sin alterar el original. También puedes especificar inplace=True en este método.

Entonces, en el caso de nuestro conjunto de datos, esta operación eliminaría 128 filas donde ingresos_millones es nulo y 64 filas donde metascore es nulo. Obviamente, esto parece un desperdicio ya que hay datos perfectamente buenos en las otras columnas de esas filas eliminadas. Es por eso que a continuación veremos la imputación.

Además de simplemente eliminar filas, también puedes eliminar columnas con valores nulos estableciendo axis=1:

```
movies_df.dropna(axis=1)
```

En nuestro conjunto de datos, esta operación eliminaría las columnas Revenue (Millions) y Metascore.

###**Imputación**

La imputación es una técnica de ingeniería de características convencional que se utiliza para conservar datos valiosos que tienen valores nulos.

Puede haber casos en los que eliminar cada fila con un valor nulo elimine una porción demasiado grande de su conjunto de datos, por lo que podemos imputar ese nulo con otro valor, generalmente la media o la mediana de esa columna.

Veamos cómo imputar los valores faltantes en la columna Revenue (Millions). Primero extraeremos esa columna en su propia variable:

```
movies_df = pd.read_csv("https://raw.githubusercontent.com/LearnDataSci/articles/master/Python%20Pandas%20Tutorial%20A%20Complete%20Introduction%20for%20Beginners/IMDB-Movie-Data.csv", index_col="Title")
revenue = movies_df['Revenue (Millions)']
revenue
```

Usar corchetes es la forma general en que seleccionamos columnas en un DataFrame.

Si recuerdas cuando creamos DataFrames desde cero, las claves del dict terminaron como nombres de columnas. Ahora, cuando seleccionamos columnas de un DataFrame, usamos corchetes como si estuviéramos accediendo a un diccionario de Python.

revenue ahora contienen una serie:

```
revenue.head()
```

Formato ligeramente diferente al de un DataFrame, pero aún tenemos nuestro índice de título.

Imputaremos los valores faltantes de ingresos utilizando la media. Aquí está el valor medio:

```
revenue_mean = revenue.mean()

revenue_mean
```

Con la media, rellenemos los valores nulos usando fillna():

```
revenue.fillna(revenue_mean, inplace=True)
```

Ahora hemos reemplazado todos los valores nulos en Revenue (Millions) por la media de la columna. Observe que al usar inplace=True en realidad hemos afectado el movie_df original:

```
movies_df.isnull().sum()
```

Imputar una columna completa con el mismo valor como este es un ejemplo básico. Sería una mejor idea intentar una imputación más granular por Genre o Director.
