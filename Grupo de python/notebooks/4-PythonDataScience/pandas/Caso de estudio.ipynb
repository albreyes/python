```python
def separate_list(nums):
  """Separates a list of integers into two lists, one for pairs and the other for odd numbers."""

  x = []
  y = []
  for i in range(len(nums)):
    if i % 2 == 0:
      x.append(nums[i])
    else:
      y.append(nums[i]*-1)

  return x, y

for l in df2.points:
  x,y = separate_list(l)
  plt.plot(x, y)
plt.show()
```
