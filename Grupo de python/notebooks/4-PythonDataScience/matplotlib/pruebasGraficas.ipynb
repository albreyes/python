```python
import matplotlib.pyplot as plt
import netCDF4
import numpy as np

ncfile = '/Users/teruhisa/mpi00/fujii/Data/osaka-bay_frc_wind_gpv_201208.nc'

nc = netCDF4.Dataset(ncfile, 'r')
#print nc.variables.keys()

time = nc.variables['wind_time'][:]


u = nc.variables['Uwind'][:,0,0]
v = nc.variables['Vwind'][:,0,0]
uv = np.sqrt(u**2 + v**2)

plt.figure(figsize=(10,2))
plt.quiver((time-61), 0, u, v, uv)
plt.xlim(22, 24)
plt.ylim(-10, 10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 7
#wind_speed = np.random.random(n)
wind_speed = np.random.uniform(0,12,n)
#wind_speed = 10
print(wind_speed)
wind_dir = np.random.uniform(0, 2*np.pi, n)
#wind_dir = np.random.uniform(0, 360, n)
print(wind_dir*180/np.pi)
time = list(range(n))
print(time)
Y = [0] * n
print(Y)

U = np.cos(wind_dir) * wind_speed
V = np.sin(wind_dir) * wind_speed

plt.figure()
plt.quiver(time, Y, U, V)
```
