## Conceptos generales

Una figura de Matplotlib la podemos descomponer en cuatro componentes principales:

- **Figure**: Se refiere a toda la figura (imagen) de nuestra visualziación. Esta figura puede tener una o más gráficas. Pensemos en un `Figure` como en un lienzo sobre el cual vamos a dibujar nuestras visualizaciones.
- **Axes**: Estos son a lo que comunmente nos referimos como *gráficas*. Un `Figure` puede contener uno o muchos `Axes`, y está por compuesto por dos o tres (en el caso de gráficas en tres dimensiones) `Axis`. Cada objeto `Axes` tiene título, etiqueta del eje x y etiqueta del eje y.
- **Axis**: Son las rectas numéricas (o rangos) encargados de generar los límites de nuestra gráfica.
- **Artist**: Todo lo que vemos en la figura es un artist. Esto incluye objetos de texto, líneas 2D, colecciones, etc.


## Pyplot

Matplotlib tiene varios módulos. Nos vamos a enfocar en Pyplot. Pyplot provee funciones simples para agregar gráficas, images, texto, etc., a los `axes` de la figura.

### Haz una gráfica simple

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.show()
```

Como podemos ver, pasamos dos listas al método `plot()` de Pyplot, y `show()` para mostrar la gráfica. La primera lista que mandamos aparece en el eje $x$, y la segunda lista parece en el eje $y$. Agreguemos título a la gráfica y etiquetas en ambos ejes. Para esto, usaremos
- `title()`
- `xlabel()`
- `ylabel()`

De la siguiente forma:

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.title("Mi primera gráfica")
plt.xlabel("Etiqueta eje x")
plt.ylabel("Etiqueta eje y")
plt.show()
```

Podemos cambiar el tamaño de la gráfica especificándolo con `figsize`

```python
plt.figure(figsize=(15,5))
plt.plot([1,2,3,4], [1,4,9,16])
plt.title("Mi primera gráfica")
plt.xlabel("Etiqueta eje x")
plt.ylabel("Etiqueta eje y")
plt.show()
```

Con las listas que mandamos como parámetros para el eje $x$ y para el eje $y$, podemos pasar un tercer argumento *opcional*. Este argumento es un string que indica el color y el tipo de línea de la gráfica. El valor por defecto es `b-`, lo cual significa `blue line`. Probemos pasando el valor `go`, o sea `green o` (círculos verdes).

```python
plt.plot([1,2,3,4], [1,4,9,16], "go")
plt.title("Mi primera gráfica")
plt.xlabel("Etiqueta eje x")
plt.ylabel("Etiqueta eje y")
plt.show()
```

Podemos graficar más de un conjunto de datos

```python
x = np.arange(1,5)
y = x**3
plt.plot([1,2,3,4], [1,4,9,16], 'go', x, y, 'r^')
plt.title("Primera gráfica")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

## Múltiples gráficas en una sola figura

Habrá veces que queramos incluir más de una gráfica en una figura. Podemos usar `subplot()` para esto.


```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[1,4,9,16], "go")
plt.title("Primera gráfica")

plt.subplot(1,2,2)
plt.plot(x,y, "y*")
plt.title("Segunda gráfica")

plt.suptitle("Múltiples gráficas")
plt.tight_layout()
plt.show()
```

 `subplot()` toma 3 parámetros: `nrows`, `ncols` e `index`, entonces realmente es `subplot(nrows, ncols, index)`
 - `nrows` indica el número de fila del subplot
 - `ncols` indica el número de columnas del subplot
 - `index` indica el índice o la posición en la que queremos colocar dicho subplot.

 En nuestro caso queríamos una figura de 1 fila con dos columnas, por lo tanto `nrows=1`, `ncols=2`.

 El primer subplot (la primera gráfica) la queremos en la posición 1 y la segunda gráfica la queremos en la posición 2. Entonces, nuestra llamada a subplot es:

 - Para la primera gráfica $\rightarrow$ `subplot(1,2,1)`
 - Para la segunda gráfica $\rightarrow$ `subplot(1,2,2)`

 Si quieramos que la figura tuviera 2 filas y una columna, entonces:
 - Para la primera gráfica $\rightarrow$ `subplot(2,1,1)`
 - Para la segunda gráfica $\rightarrow$ `subplot(2,1,2)`

```python
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[1,4,9,16], "go")
plt.title("Primera gráfica")

plt.subplot(2,1,2)
plt.plot(x,y, "y*")
plt.title("Segunda gráfica")

plt.suptitle("Múltiples gráficas")
plt.show()
```

Este método se vuelve un poco tedioso cuando queremos muchos subplots en una misma figura. Una forma más conveniente y sencilla es utilizando la función `subplots()`. Este método recibe dos parámetros: `nrows` $\rightarrow$ número de renglones, y `ncols` para número de columnas. Al ejecutar este método, se crearán dos objetos: `figure` y `axes` (en manera de tupla).

```python
x = np.arange(1,10)
y = x**3
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
ax[0,1].plot([1,2,3,4], [1,4,9,16], 'go')
ax[1,0].plot(x,y,"r^")
ax[0,1].set_title("Cuadrados")
ax[1,0].set_title("Cubos")
plt.tight_layout() #par mejor acomodo
plt.show()
```

## Diferentes tipos de gráficas

## Gráficas de barras

Las gráficas de barra son sin duda de las más utilizadas en el campo de visualización de datos. Permiten fáclmente asociar datos a variables categóricas. Por suerte, `pyplot` provee un método `bar()` para crear gráficas de barras especificando (como parámetros) las variables categóricas, sus valores y opcionalmente el color.

```python
grupos = ["A","B","C","D","E"]
calificacion_promedio_grupo = [70,82,73,65,68]

plt.bar(grupos, calificacion_promedio_grupo)
plt.title("Gráfica de barras")
plt.xlabel("Grupo")
plt.ylabel("Promedio")
plt.show()
```

Especificando color verde

```python
grupos = ["A","B","C","D","E"]
calificacion_promedio_grupo = [70,82,73,65,68]

plt.bar(grupos, calificacion_promedio_grupo, color="green")
plt.title("Gráfica de barras")
plt.xlabel("Grupo")
plt.ylabel("Promedio")
plt.show()
```

Tambien podemos hacer la gráfica con barras horizontales utilizando `barh()`

```python
grupos = ["A","B","C","D","E"]
calificacion_promedio_grupo = [70,82,73,65,68]

plt.barh(grupos, calificacion_promedio_grupo, color="red")
plt.title("Gráfica de barras")
plt.xlabel("Promedio")
plt.ylabel("Grupo")
plt.show()
```

## Gráfica de barras con dos variables

```python
grupos = ["A","B","C","D","E"]
calificacion_promedio_grupo = [70,82,73,65,68]
calificacion_promedio_hombres = [68,67,77,61,70]


indice = np.arange(5)
width=0.3


plt.bar(indice, calificacion_promedio_grupo, width, color="red")
plt.bar(indice+width, calificacion_promedio_hombres, width, color="green")

plt.show()
```

Mejorando la gráfica de arriba

```python
grupos = ["A","B","C","D","E"]
calificacion_promedio_grupo = [70,82,73,65,68]
calificacion_promedio_hombres = [68,67,77,61,70]


indice = np.arange(5)
width=0.3

plt.bar(indice, calificacion_promedio_grupo, width, color="red")
plt.bar(indice+width, calificacion_promedio_hombres, width, color="green")

#agregamos título
plt.title("Gráfica de dos barras")
plt.xlabel("Grupos")
plt.ylabel("Calificación")
plt.show()
```

Todavía aparece el eje X con los valores `[0,1,2,3,4]`. Se vería mejor si tuvieramos en el eje X a cada grupo. Para esto podemos utilizar `xticks()`.

```python
grupos = ["A","B","C","D","E"]
calificacion_promedio_grupo = [70,82,73,65,68]
calificacion_promedio_hombres = [68,67,77,61,70]


indice = np.arange(5)
width=0.3


plt.bar(indice, calificacion_promedio_grupo, width, color="red")
plt.bar(indice+width, calificacion_promedio_hombres, width, color="green")

#agregamos xticks
plt.xticks(indice + width/2, grupos)

plt.title("Gráfia de dos barras")
plt.xlabel("Grupos")
plt.ylabel("Calificación")
plt.show()
```

Por último queremos agregar una leyenda a la gráfica para saber qué variable está representada en cada color. Para esto, especificaremos un parámetro `label` en cada barra, y agregaremos un valor de `legend()` a nuestra figura

```python
grupos = ["A","B","C","D","E"]
calificacion_promedio_grupo = [70,82,73,65,68]
calificacion_promedio_hombres = [68,67,77,61,70]


indice = np.arange(5)
width=0.3

                                                                           #Agregamos label a cada barra
plt.bar(indice, calificacion_promedio_grupo, width, color="red",           label="Promedio Grupo")
plt.bar(indice+width, calificacion_promedio_hombres, width, color="green", label="Promedio Hombres")


plt.xticks(indice + width/2, grupos)


plt.title("Gráfia de dos barras")
plt.xlabel("Grupos")
plt.ylabel("Calificación")

#especificamos que coloque la leyenda en donde mejor quede en la figuraÇ
plt.legend(loc='best')

plt.show()
```

## Gráficas de pastel (pie charts)

```python
candidatos = ["Anaya", "Meade", "AMLO", "Bronco"]
porcentaje = [22.2750,16.4099,53.1936,5.2317]
colores = ["blue", "green", "brown", "purple"]
extruir = [0,0,0.1,0]

plt.pie(porcentaje, explode=extruir, labels=candidatos, shadow=True, startangle=45, colors=colores)
plt.axis('equal')
#plt.axis('scaled')
plt.legend(title="Candidatos")
plt.show()
```

## Histogramas

Los histogramas son la opción más adecuada cuando tenemos variables continuas. Ejemplos de variables continuas: altura, peso, precios, tiempos de espera, etc.

La visualización de un histograma consiste en la frecuencia (la cuenta total) de elementos que 'caen' dentro de un rango.

A continuación generaremos 1000 datos aleatorios (pseudoaleatorios porque fijaremos una semilla) de una distribución normal estándar, y posteriormente graficaremos estos datos en un histograma que divida a los datos en 10 'cubetas'.

```python
np.random.seed(1234)
x = np.random.randn(1000)

plt.title("Histograma")
plt.xlabel("Datos aleatorios")
plt.ylabel("Frecuencia")
plt.hist(x,10)
plt.show()
```

```python
np.random.seed(1234)
x = np.random.randn(10000)

plt.title("Histograma")
plt.xlabel("Datos aleatorios")
plt.ylabel("Frecuencia")
plt.hist(x,100)
plt.show()
```

```python
alturas = [167,170,149,165,155,180,166,146,159,185,145,168,172,181,169]
pesos = [86,74,66,78,68,79,90,73,70,88,66,84,67,84,77]

plt.xlim(140,200)
plt.ylim(60,100)
plt.scatter(alturas, pesos)
plt.title("Gráfica de dispersión")
plt.xlabel("Altura (cm)")
plt.ylabel("Peso (kg)")
```

## Resumen

- `plot(valores_eje_x, valores_eje_y)` — gráfica de línea de valores de x en y
- `show()` — Mostrar la gráfica
- `title("string")` — setear el título de la gráfica
- `xlabel("string")` — setear la etiqueta del eje x
- `ylabel("string")` — setear la etiqueta del eje y
- `figure()` — controlar los atributos a nivel figura
- `subplot(nrows, ncols, index)` — Agregar un subplot a la figura actual
- `suptitle("string")` — Agregar un título que compartan todos los subplots
- `subplots(nrows, ncols, figsize)` — Una forma más fácil de crear múltiples gráficas en una sola figura. Nos regresa una tupla de una figura y número de axes.
- `set_title("string")` — método a nivel de las axes para color título de un subplot
- `bar(variables_categoricas, valores, color)` — crear gráficas de barras
- `barh(variables_categoricas, valores, color)` — crear gráficas de barras horizontales
- `legend(loc)` — agregar la leyenda de la gráfica
- `xticks(indice, variables_categoricas)` — Setear la ubicación y las etiquetas del eje x
- `pie(valores, variables_categoricas)` — crear gráfica de pay
- `hist(valores, numero_cubetas)` — crear histogramas
- `xlim(valor_inicial, valor_final)` — setear el valor máximo de las x
- `ylim(valor_inicial, valor_final)` — setear el valor máximo de las y
- `scatter(valores_eje_x, valores_eje_y)` — crear gráfia de dispersión
- `axes()` — agregar axes a la figura actual
- `set_xlabel("string")` — método a nivel de las axes para color la etiqueta del eje x de la figura actual
- `set_ylabel("string")` — método a nivel de las axes para color la etiqueta del eje y de la figura actual
