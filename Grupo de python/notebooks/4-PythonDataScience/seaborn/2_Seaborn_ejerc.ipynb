#**Ejercicio: Visualización de datos con seaborn**

En este ejercicio usaremos seaborn para hacer análisis visual en una lista Billboard Hot 100 del año 2000.

La lista Billboard Hot 100 contiene información estándar de la industria musical en los Estados Unidos para canciones, publicada semanalmente por la revista Billboard. Las clasificaciones de las listas se basan en las ventas, la reproducción de radio y la transmisión en línea en los Estados Unidos.

Cada semana, Billboard publica la lista "The Hot 100" de canciones que fueron tendencia en ventas y difusión durante esa semana. Este conjunto de datos es una colección de todos los gráficos "The Hot 100" publicados desde su creación en 1958.

Los datos del chart incluyen clasificaciones de las canciones en el tiempo.

Ejemplo obtenido desde [Billboard data](https://ift6758.github.io/lectures/billboard.html)

```python
url='https://raw.githubusercontent.com/albreyes/python/main/Grupo%20de%20python/databases/billboard.csv'
import pandas as pd
datos=pd.read_csv(url)
```

```python
# eliminamos las columnas con datos nulos (NAN o null)
datos_sin_nulos=datos.dropna(axis=1)
```

```python
#revisaremos si hay registros duplicados
temporal=datos_sin_nulos.drop_duplicates()
```

```python
# elimnamos la columna year que siempre tiene el mis dato
temporal=datos_sin_nulos.drop(columns = ["year"])
temporal
```

Ahora si usamos *seaborn* para hacer el anlisis visual de nuestro conjunto de datos. Primero, hagamos un análisis de frecuencias por registro ...

```python
import seaborn as sb
#sb.factorplot('genre',data=temporal,kind="count")
#sb.displot(temporal['x1st.week'])
sb.histplot(temporal['x1st.week'])
```

```python
sb.countplot(x='genre',data=temporal);
```

```python
#sb.countplot(y='artist.inverted',data=temporal);
sb.countplot(y='genre',data=temporal);
```
